## **Zadatak 1: Izračun deskriptivne statistike i testiranje distribucija**

Za numeričke varijable u skupu podataka izračunajte osnovne mjere u okviru deskriptivne statistike. Zatim se fokusirajte na dohodak po stanovniku države (*IncomePerCap*). Koliko u prosjeku zarađuju stanovnici SAD-a? Kolika je medijalna plaća? 

```python
num_cols = df.select_dtypes(include=np.number).columns

display(df[num_cols].describe().T)

income = df["IncomePerCap"].dropna()

print("IncomePerCap")
print(f"Prosjek:  {income.mean():.2f}")
print(f"Medijan:  {income.median():.2f}")
print(f"Std dev:  {income.std():.2f}")
print(f"Min:      {income.min():.2f}")
print(f"Q1:       {income.quantile(0.25):.2f}")
print(f"Q3:       {income.quantile(0.75):.2f}")
print(f"Max:      {income.max():.2f}")
```

Pokušajte steći dublji uvid u prihod po stanovniku. Može li se prepoznati distribucija iz koje potječi ovi podaci? Potkrijepite Vaš zaključak grafičkom metodom i rezultatima statističkog testa.

```python
income = df["IncomePerCap"].dropna()

fig, ax = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(income, kde=True, ax=ax[0])
ax[0].set_title("Histogram IncomePerCap")

stats.probplot(income, dist="norm", plot=ax[1])
ax[1].set_title("Q-Q plot za normalnu distribuciju")

plt.tight_layout()
plt.show()

z_income = (income - income.mean()) / income.std(ddof=1)

dagostino_stat, dagostino_p = stats.normaltest(income)
ks_stat, ks_p = stats.kstest(z_income, "norm")

print(f"D'Agostino test: statistika={dagostino_stat:.4f}, p-vrijednost={dagostino_p:.4e}")
print(f"Kolmogorov-Smirnov test na standardiziranim podacima: statistika={ks_stat:.4f}, p-vrijednost={ks_p:.4e}")
```

Izračunajte osnovne mjere u okviru deskriptivne statistike za broj stanovnika države (*TotalPop*) te prokomentirajte dobivene rezultate.

```python
pop = df["TotalPop"].dropna()

print("TotalPop")
print(f"Prosjek:  {pop.mean():.2f}")
print(f"Medijan:  {pop.median():.2f}")
print(f"Std dev:  {pop.std():.2f}")
print(f"Min:      {pop.min():.2f}")
print(f"Q1:       {pop.quantile(0.25):.2f}")
print(f"Q3:       {pop.quantile(0.75):.2f}")
print(f"Max:      {pop.max():.2f}")
print(f"Skewness: {stats.skew(pop):.4f}")
print(f"Kurtosis: {stats.kurtosis(pop):.4f}")
```

Pokušajte steći dublji uvid u broj stanovnika. Može li se prepoznati distribucija iz koje potječu ovi podaci? Potkrijepite Vaš zaključak grafičkom metodom i rezultatima statističkog testa.

```python
pop = df["TotalPop"].dropna()

fig, ax = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(pop, kde=True, ax=ax[0])
ax[0].set_title("Histogram TotalPop")

stats.probplot(pop, dist="norm", plot=ax[1])
ax[1].set_title("Q-Q plot za normalnu distribuciju")

plt.tight_layout()
plt.show()

z_pop = (pop - pop.mean()) / pop.std(ddof=1)

dagostino_stat, dagostino_p = stats.normaltest(pop)
ks_stat, ks_p = stats.kstest(z_pop, "norm")

print(f"D'Agostino test: statistika={dagostino_stat:.4f}, p-vrijednost={dagostino_p:.4e}")
print(f"Kolmogorov-Smirnov test na standardiziranim podacima: statistika={ks_stat:.4f}, p-vrijednost={ks_p:.4e}")
```

## **Zadatak 2: Ispitivanje odnosa između dvije varijable**

Pokušajte odgovoriti na pitanje je li dohodak po stanovniku države u korelaciji s nezaposlenošću. Ispitajte i linearnu i monotonu korelaciju.

```python
corr_df = df[["IncomePerCap", "Unemployment"]].dropna()

pearson_r, pearson_p = stats.pearsonr(corr_df["IncomePerCap"], corr_df["Unemployment"])
spearman_rho, spearman_p = stats.spearmanr(corr_df["IncomePerCap"], corr_df["Unemployment"])

print(f"Pearson r = {pearson_r:.4f}, p-vrijednost = {pearson_p:.4e}")
print(f"Spearman rho = {spearman_rho:.4f}, p-vrijednost = {spearman_p:.4e}")

plt.figure(figsize=(7, 5))
sns.regplot(data=corr_df, x="Unemployment", y="IncomePerCap", scatter_kws={"alpha": 0.5})
plt.title("IncomePerCap vs Unemployment")
plt.show()
```

## **Zadatak 3: Testiranje hipoteza**

Zarađuju li stanovnici države New York više od stanovnika države Kalifornija? Pokušajte dati odgovor na ovo pitanje pomoću deskriptivne statistike.

```python
ny = df.loc[df["State"] == "New York", "IncomePerCap"].dropna()
ca = df.loc[df["State"] == "California", "IncomePerCap"].dropna()

summary = pd.DataFrame({
    "New York": [ny.count(), ny.mean(), ny.median(), ny.std()],
    "California": [ca.count(), ca.mean(), ca.median(), ca.std()]
}, index=["n", "mean", "median", "std"])

display(summary)
print(f"Razlika sredina (NY - CA): {ny.mean() - ca.mean():.2f}")
```

Kako bi potkrijepili dobivene zaključke, provedite testiranje hipoteza. Neka nulta hipoteza pretpostavlja da nema statistički značajne razlike u dohotku stanovnika ovih država. Testirajte hipotezu pomoću dvostranog t-testa (*two-sided t-test*). Prokomentirajte dobivenu $p$-vrijednost te donesite zaključak.

```python
ttest = stats.ttest_ind(ny, ca, equal_var=False)  # Welch t-test

print(f"t-statistika = {ttest.statistic:.4f}")
print(f"p-vrijednost = {ttest.pvalue:.4e}")
```

## **Zadatak 4: Kvantificiranje nesigurnosti**

Ponovno promotrimo pitanje zarađuju li stanovnici države New York više od stanovnika države Kalifornija. Ovaj put pokušajte odgovoriti na pitanje pomoću intervala pouzdanosti s razinom pouzdanosti od 95%. Uz izračun, vizualizirajte intervale pouzdanosti na stupčastim dijagramima. 
Pomoć: provjerite preklapaju li se intervali za ove države.

```python
def mean_ci(x, confidence=0.95):
    x = pd.Series(x).dropna()
    n = len(x)
    mean = x.mean()
    sem = stats.sem(x)
    lo, hi = stats.t.interval(confidence, df=n-1, loc=mean, scale=sem)
    return mean, lo, hi

ny_mean, ny_lo, ny_hi = mean_ci(ny)
ca_mean, ca_lo, ca_hi = mean_ci(ca)

ci_df = pd.DataFrame({
    "State": ["New York", "California"],
    "MeanIncomePerCap": [ny_mean, ca_mean],
    "CI_low": [ny_lo, ca_lo],
    "CI_high": [ny_hi, ca_hi]
})
ci_df["err_low"] = ci_df["MeanIncomePerCap"] - ci_df["CI_low"]
ci_df["err_high"] = ci_df["CI_high"] - ci_df["MeanIncomePerCap"]

display(ci_df[["State", "MeanIncomePerCap", "CI_low", "CI_high"]])

plt.figure(figsize=(8, 5))
plt.bar(ci_df["State"], ci_df["MeanIncomePerCap"],
        yerr=[ci_df["err_low"], ci_df["err_high"]], capsize=8)
plt.ylabel("IncomePerCap")
plt.title("95% intervali pouzdanosti za IncomePerCap")
plt.show()
```

## **Zadatak 5: Jačanje gospodarstva poticanjem samozapošljavanja**

Pretpostavimo da je najveći prioritet lokalne uprave u 2018. godini povećanje dohotka po glavi stanovnika. U tu svrhu, dužnosnici planiraju usvojiti strategiju poticanja samozapošljavanja kroz niz kampanja, obrazovnih programa i namjenskih sredstava.

Budući da je neetično i nemoguće u ovom okruženju provoditi kontrolirani eksperiment koji uključuje građane (primjerice, otpustiti zaposlenike i prisiliti ih na samozapošljavanje), dužnosnici su od Vas zatražili da utvrdite učinak samozapošljavanja na gospodarstva, oslanjajući se na podatke iz danog skupa podataka.

Koristeći odgovarajuće grafičke i statističke metode, odredite i kvantificirajte vezu između postotka samozaposlenih građana (*SelfEmployed*) i dohotka po glavi stanovnika u svih 3212 država SAD-a. Zarađuju li građani u državama koje imaju veći postotak samozaposlenih više po glavi stanovnika?

```python
se_df = df[["SelfEmployed", "IncomePerCap"]].dropna()

pearson_r, pearson_p = stats.pearsonr(se_df["SelfEmployed"], se_df["IncomePerCap"])
spearman_rho, spearman_p = stats.spearmanr(se_df["SelfEmployed"], se_df["IncomePerCap"])
slope, intercept, r_value, p_value, std_err = stats.linregress(se_df["SelfEmployed"], se_df["IncomePerCap"])

print(f"Pearson r = {pearson_r:.4f}, p-vrijednost = {pearson_p:.4e}")
print(f"Spearman rho = {spearman_rho:.4f}, p-vrijednost = {spearman_p:.4e}")
print(f"Linearna regresija: IncomePerCap = {intercept:.2f} + {slope:.2f} * SelfEmployed")
print(f"R^2 = {r_value**2:.4f}")

plt.figure(figsize=(7, 5))
sns.regplot(data=se_df, x="SelfEmployed", y="IncomePerCap", scatter_kws={"alpha": 0.35})
plt.title("IncomePerCap vs SelfEmployed (svi okruzi)")
plt.show()
```

Pilot program ove strategije zahvatit će sve okruge unutar ograničenog skupa od tri države. Skup A uključuje Wisconsin, Tennessee i Minnesotu. Kvantificirajte ovisnost dohotka po glavi stanovnika o stopama samozapošljavanja u skupu A.

```python
states_A = ["Wisconsin", "Tennessee", "Minnesota"]
df_A = df[df["State"].isin(states_A)].copy()

plt.figure(figsize=(8, 6))
sns.lmplot(data=df_A, x="SelfEmployed", y="IncomePerCap", hue="State", height=6, aspect=1.2)
plt.show()

results_A = []
for state in states_A:
    tmp = df_A[df_A["State"] == state][["SelfEmployed", "IncomePerCap"]].dropna()
    pearson_r, pearson_p = stats.pearsonr(tmp["SelfEmployed"], tmp["IncomePerCap"])
    spearman_rho, spearman_p = stats.spearmanr(tmp["SelfEmployed"], tmp["IncomePerCap"])
    slope, intercept, r_value, p_value, std_err = stats.linregress(tmp["SelfEmployed"], tmp["IncomePerCap"])
    results_A.append([state, len(tmp), pearson_r, pearson_p, spearman_rho, spearman_p, slope, r_value**2])

results_A = pd.DataFrame(
    results_A,
    columns=["State", "n", "Pearson_r", "Pearson_p", "Spearman_rho", "Spearman_p", "Slope", "R2"]
)
display(results_A.sort_values("Slope", ascending=False))
```

U kojoj je državi unutar skupa A opaženi učinak samozapošljavanja na dohodak po glavi stanovnika najjači? Je li ovaj učinak pozitivan ili negativan?

```python
strongest_A = results_A.loc[results_A["Slope"].abs().idxmax()]

print("Najjači učinak u skupu A:")
print(strongest_A)

if strongest_A["Slope"] > 0:
    print("Učinak je pozitivan.")
else:
    print("Učinak je negativan.")
```

Skup B uključuje države New Jersey, Kansas i Rhode Island. Ponovite analizu iz prijašnjih koraka, ali sada za skup B. U kojem biste od dva skupa A i B preporučili poticanje samozapošljavanja? Objasnite svoje razmišljanje. Što zamjećujete kad u oba skupa promatrate korelaciju unutar svake pojedine države, a što kada promatrate korelaciju za države u cjelini?

```python
states_B = ["New Jersey", "Kansas", "Rhode Island"]
df_B = df[df["State"].isin(states_B)].copy()

plt.figure(figsize=(8, 6))
sns.lmplot(data=df_B, x="SelfEmployed", y="IncomePerCap", hue="State", height=6, aspect=1.2)
plt.show()

results_B = []
for state in states_B:
    tmp = df_B[df_B["State"] == state][["SelfEmployed", "IncomePerCap"]].dropna()
    pearson_r, pearson_p = stats.pearsonr(tmp["SelfEmployed"], tmp["IncomePerCap"])
    spearman_rho, spearman_p = stats.spearmanr(tmp["SelfEmployed"], tmp["IncomePerCap"])
    slope, intercept, r_value, p_value, std_err = stats.linregress(tmp["SelfEmployed"], tmp["IncomePerCap"])
    results_B.append([state, len(tmp), pearson_r, pearson_p, spearman_rho, spearman_p, slope, r_value**2])

results_B = pd.DataFrame(
    results_B,
    columns=["State", "n", "Pearson_r", "Pearson_p", "Spearman_rho", "Spearman_p", "Slope", "R2"]
)
display(results_B.sort_values("Slope", ascending=False))

# agregirano po skupovima
def group_summary(subdf, label):
    tmp = subdf[["SelfEmployed", "IncomePerCap"]].dropna()
    pearson_r, pearson_p = stats.pearsonr(tmp["SelfEmployed"], tmp["IncomePerCap"])
    spearman_rho, spearman_p = stats.spearmanr(tmp["SelfEmployed"], tmp["IncomePerCap"])
    slope, intercept, r_value, p_value, std_err = stats.linregress(tmp["SelfEmployed"], tmp["IncomePerCap"])
    return pd.Series({
        "Group": label,
        "n": len(tmp),
        "Pearson_r": pearson_r,
        "Pearson_p": pearson_p,
        "Spearman_rho": spearman_rho,
        "Spearman_p": spearman_p,
        "Slope": slope,
        "R2": r_value**2
    })

comparison = pd.DataFrame([
    group_summary(df_A, "A"),
    group_summary(df_B, "B")
])

display(comparison)
```
